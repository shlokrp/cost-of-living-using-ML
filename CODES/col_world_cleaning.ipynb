# Filling missing values by taking average of inflation rates of all the years for a particular country

```python
# Calculate the average inflation rate for each country across all years
average_inflation_by_country = df_inflation2.loc[:, '2000':'2022'].mean(axis=1)

# Define a function to fill missing values in each row with the country's average inflation rate
def fill_missing_values(row):
    country = row['country']
    avg_inflation = average_inflation_by_country.loc[row.name]  # Get the average inflation rate for the country
    return row.fillna(avg_inflation)

# Apply the fill_missing_values function to each row
df_inflation2.loc[:, '2000':'2022'] = df_inflation2.apply(fill_missing_values, axis=1)

# Check if there are any remaining null values
remaining_null_values = df_inflation2.isnull().sum().sum()
if remaining_null_values == 0:
    print("All null values have been filled with the average inflation rate for each country.")
else:
    print(f"There are still {remaining_null_values} null values remaining after filling.")
```
