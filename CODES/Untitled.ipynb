```python
import matplotlib.pyplot as plt

# Function to plot the architecture of Exponential Smoothing
def plot_exponential_smoothing_architecture():
    # Define the layers of the architecture
    layers = ['Input', 'Exponential Smoothing', 'Output']

    # Define the connections between layers
    connections = [('Input', 'Exponential Smoothing'), ('Exponential Smoothing', 'Output')]

    # Create the plot
    plt.figure(figsize=(8, 4))
    
    # Plot the layers
    for layer in layers:
        plt.plot(layer, 'o', markersize=10, label=layer)

    # Plot the connections
    for connection in connections:
        plt.plot(connection, marker='o', linestyle='--', markersize=0, linewidth=2, color='gray')

    # Add labels and title
    plt.title('Architecture of Exponential Smoothing')
    plt.xlabel('Layers')
    plt.ylabel('Components')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Call the function to plot the architecture
plot_exponential_smoothing_architecture()
```
